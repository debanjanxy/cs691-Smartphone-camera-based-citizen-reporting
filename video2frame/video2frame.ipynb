```python
import cv2
import numpy as np
import os
```

```python
filename = input("Enter file name :")
videoname = input("Enter name of the video: ")
try:
    if not os.path.exists(filename):
        os.makedirs(filename)
except OSError:
    print("Error: Creating directory named ", filename)
```

```python
currentFrame = 0
cap = cv2.VideoCapture(videoname)
frame = np.zeros(shape=(1080,1920))
while(frame is not None):
    ret,frame = cap.read()
    name = './'+filename+'/frame'+str(currentFrame)+'.jpg'
    print("Creating..."+name)
    cv2.imwrite(name,frame)
    currentFrame += 1

print("Task Successful!")
cap.release()
cv2.destroyAllWindows()
```
